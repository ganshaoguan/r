## 编写R函数

R函数的基本格式为 functionname <- function(a1, a2, ...) expression

```r
# 编写一个计算平均数、中位数、极差和标准差的函数
load("example1_1.RData")
head(example1_1)
x <- example1_1[,4]   # 把数值赋值给x
```

```r
myfun <- function(x){
    n = length(x)
    mean = sum(x)/n
    median = median(x)
    r = max(x) - min(x)
    s = sd(x)
    summ  = data.frame(c(mean, median,r, s),
                      row.names = c("均值","中位数","极差","标准差") 
                      )
    names(summ) = "值"
    return(summ)
}
```

```r
myfun(x)
```

### 数值数据类别化

对数值数据，可以通过将数据分组实现类别化

对于一组数值数据，要进行分组，首先要确定组数，其次确定组距；
- 根据Sturges提出的组数确定方法，设组数为k，则 $k =1 + \frac{lg(n)}{lg(2)}$ ,在R中可使用函数 nclass.Sturges(x)来计算；
- 组距可根据极差和组数来确定，即 组距 = (最大值 - 最小值) / k

```r
getwd()
setwd("D:/statistics/jjp/data_jiajp/example/ch2")
load("example2_2.RData")
```

```r
head(example2_2, 3)
nclass.Sturges(example2_2$`销售额`)  # 这里计算的组数k = 8，但书上为便于理解，将组数设定为12
# (max(example2_2$`销售额`) - min(example2_2$`销售额`))/8  # 这是理论的组距
(max(example2_2$`销售额`) - min(example2_2$`销售额`))/12  # 组数为12时，组距为9.25，为便于计算，可将组距设置为5或者10
```

```r
# 数据分组与生成频数表
vector2_2 <- as.vector(example2_2$`销售额`)   # 转换成向量
library(plyr)
count <- table(round_any(vector2_2, 10, floor))
# round_any() 四舍五入到任意数的倍数
# > round_any(135, 10, ceiling)  #ceiling 向上取整，倍数为10
# [1] 140 # 向上最接近135且是10的倍数的数字
# > round_any(135, 10, floor) #floor 向下取整，倍数为10
# [1] 130 #向下最接近135且是10的倍数的数字
# > round_any(135, 25, ceiling)  #向上取整，倍数为25
# [1] 150 # 向上最接近135且是25得倍数得数字
# > round_any(135, 25, floor) #向上取整，倍数为25
# [1] 125 # 向下最接近135且是25得倍数得数字

count <- table(round_any(vector2_2, 10, floor))
count <- as.numeric(count)
pcount <- prop.table(count) * 100
cumsump <- cumsum(pcount)   # 累计求和
name <- paste(seq(160, 270, by = 10),'-', seq(170, 280, by = 10), sep = "")
gt <- data.frame("频数" = count, "百分比" = pcount, "累计百分比" = cumsump, row.names = name)
round(gt, 2)  # 保留了两位小数
```

#### 箱线图

箱线图可反应一组数据的分布特征，更多是用再比较多组数据的分布特征；
箱线图主要包括的元素有：

    中位数
    上四分位数
    下四分位数
    四分位差(IQR) = 上四分位数-下四分位数
    上内围栏 = 上四分位数 + 1.5 * IQR
    下内围栏 = 下四分位数 - 1.5 * IQR
    上下内围栏之内的最大值和最小值称为上相邻值和下相邻值
    上下内围栏之外的数据则称之为离群点
    
需要注意的是，也可以设置3倍的IQR作为上线围栏的界限； 图示如下



```r
# 绘制不同设置的箱线图
par(mfrow = c(3,1),  # 生成1行3列的画布
   mai = c(0.7, 0.7, 0.6, 0.1),
    cex = 1.3,    # 图形文字及字符大小
    cex.main = 1.8   # 主题文字大小
   )

x <- rnorm(1000, 50 ,5)

# 相邻值与箱子连线的箱线图

boxplot(x,
       range = 1.5,  # 设置1.5倍的IQR
        col = 'red',
        horizontal = TRUE,   # 水平放置 = TRUE
        main = "相邻值与箱子连线的箱线图(range = 1.5)",
        cex = 0.8
       )

boxplot(x, 
       range = 3,
        col = 'green',
        horizontal= T,
        main  ="相邻值与箱子连线的箱线图(range = 3)",
        cex = 0.8
       )

#极值与箱子连线的箱线图

boxplot(x, range = 0,
        varwidth = T,
        col = 'pink',
        horizontal = T,
        main = "极值与箱子连线的箱线图(range=0, varwidth = T)",
        cex = 0.8
       )

# boxplot函数的参数解析
# boxplot(x,
#     range = 1.5,  # IQR
#     width = NULL, # 自定义箱子的宽度
#         varwidth = TRUE, 箱子的宽度与样本量的平方根呈比例
#         notch = TRUE, 可绘制凹槽箱线图
#         horizontal = TRUE, 绘制水平箱线图，就是将箱线图横着放
# )
```

### 洛伦茨曲线

在频数分布中，如果将各类别的频数逐级累加，即可得到累积频数分布表。根据累加频数分布表可以绘制累加频数分布曲线。


**洛伦茨曲线**（Lorenz curve）是一种特殊的理解频数分布曲线，它是20世纪初由美国统计学家洛伦茨(M. E. Lorentz)根据意大利经济学家帕累托(V. Pareto)提出的收入分配公式绘制的**描述收入和财富分配不平等程度的曲线**

洛伦茨曲线的一般形式如下

图中弯曲的线就是洛伦茨曲线  
如果用横轴表示人口百分比，纵轴表示相应人口获得的收入百分比，通过洛伦兹曲线，可以直观地反映一个国家或地区收入分配平等或不平等的状况  
如果一定累积百分比的人口获得相同累积百分比的收入，就是图中的对角线，即收入分配绝对平均线。  
如果绝大多数人口占有很少的收入，而一小部分人口占有绝大部分的收入，则洛伦茨曲线就靠近下横轴和右纵轴形成弯曲的线。弯曲程度越大，表示收入分配越不公平


**基尼系数**  
$$基尼系数 = \frac{A}{A+B}$$

A表示实际收入洛伦茨曲线与绝对平均线之间的面积，B表示实际收入洛伦茨曲线与绝对不平均线之间的面积。  
如果A = 0，则基尼系数= 0，表示收入绝对平均；如果B = 0，则基尼系数= 1，表示收入绝对不平均。  
基尼系数通常在0 和1之间取值，数值越小，表示收入分配越平均，数值越大表示收入分配越不平均。  
一般认为，基尼系数若小于0.2，表明收入分配平均；基尼系数在0.2～0.4之间表示收入分配比较合理；基尼系数为0.4时，被认为是收入分配不公平的警戒线，超过了0.4表示收入差距较大

```r
# 绘制洛伦茨曲线
getwd()
setwd('D:/statistics/jjp/data_jiajp/example/ch2')
load("example2_10.RData")
d <- example2_10
d

# 根据已有数据的形式，需要计算出各组的人数占比，累计占比，收入占比， 收入累计占比
```
